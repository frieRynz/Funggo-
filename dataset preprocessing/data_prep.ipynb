```python
import numpy as np
import pandas as pd
import re
import os
```

```python
dataset_path = 'PUT YOUR DATASET PATH HERE'
output_path = 'PUT YOUR OUTPUT PATH HERE'
```

```python
chunksize = 50000

filtered_chunks = []
for chunk in pd.read_csv(dataset_path, chunksize=chunksize):
    chunk = chunk[chunk['language'].isin(['en', 'th'])]
    filtered_chunks.append(chunk)

df = pd.concat(filtered_chunks)
```

```python
df = df[(df['year'] >= 1990) & (df['year'] <= 2022)]
```

```python
keep_cols = ["title", "artist", "year", "views", "lyrics"]
drop_cols = df.columns.difference(keep_cols)

df = df.drop(columns=drop_cols)
```

```python
# pattern to detect translations (in both EN + TH)
bad_pattern = re.compile(
    r'(แปลไทย|แปลภาษาไทย|translation|translated|เวอร์ชันไทย|thai version|english translation)',
    re.IGNORECASE
)

# keep only rows that DO NOT match
df = df[~df['title'].str.contains(bad_pattern, na=False)]
```

```python
df.to_csv("dataset/genius_for_search.csv", index=False)
```
